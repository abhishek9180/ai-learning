# 📓 Step 4: First AI Notebook

Create a new notebook:
`01_basic_chat.ipynb`

## 🧪 Code Cell

```python
import requests

OLLAMA_URL = "http://localhost:11434/api/generate"

def ask_llm(prompt):
    response = requests.post(
        OLLAMA_URL,
        json={
            "model": "mistral",  # or "phi"
            "prompt": prompt,
            "stream": False
        }
    )
    
    return response.json()["response"]

# Test
response = ask_llm("Explain recursion in simple terms")
print(response)
```

# One-time prompt example

````python
import requests
import json

OLLAMA_URL = "http://localhost:11434/api/generate"

def ask_llm(prompt):
    response = requests.post(
        OLLAMA_URL,
        json={
            "model": "mistral",  # or "phi"
            "prompt": prompt,
            "stream": False
        }
    )
    
    return response.json()["response"]

# Test
response = ask_llm("Explain recursion in simple terms")
print(response)
````

# Interactive Chat

````python
while True:
    user_input = input("You: ")
    
    if user_input.lower() in ["exit", "quit"]:
        break
    
    response = ask_llm(user_input)
    print("AI:", response)
````
