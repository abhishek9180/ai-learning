# 🧠 Module 1: Prompt Engineering & Controlled Outputs

---

## 🎯 Objective
Learn how to control LLM responses using structured prompting and enforce reliable outputs.


## ⚙️ Setup
(Import libraries and define helper functions)

```python
import requests
import json

OLLAMA_URL = "http://localhost:11434/api/generate"

def ask_llm(prompt, model="mistral"):
    response = requests.post(
        OLLAMA_URL,
        json={
            "model": model,
            "prompt": prompt,
            "stream": False
        }
    )
    return response.json()["response"]
```

## 🧪 Experiment 1: Basic Prompt vs Structured Prompt

### 🔹 Simple Prompt
(Free-form response)

```python
prompt = "Extract name and age from: John is 25 years old"
print(ask_llm(prompt))
```

### 🔹 Structured Prompt
(Control tone, format)

```python
prompt = """
Extract the following fields:
- name
- age

Return ONLY valid JSON.

Text:
John is 25 years old
"""

print(ask_llm(prompt))
```

### 📌 Observation
Small prompt changes = huge output difference.


## 🧪 Experiment 2: System Prompt Behavior


```python

prompt = """
You are a strict assistant.
You ONLY respond in bullet points.
No extra explanation.

Question: What is Python?
"""

print(ask_llm(prompt))
```

## 🧪 Experiment 3: JSON Output Enforcement

```python
# Prompt for JSON Output

prompt = """
Extract information in JSON format with keys:
name, age

Rules:
- Output MUST be valid JSON
- No explanation
- No extra text

Text:
Alice is 30 years old
"""

response = ask_llm(prompt)
print(response)
```

```python
# Parse JSON

try:
    data = json.loads(response)
    print("Parsed:", data)
except:
    print("Invalid JSON:", response)
```

## 🧪 Experiment 4: Handling Broken Outputs

```python
# Add Retry Logic
def ask_json(prompt, retries=3):
    for _ in range(retries):
        response = ask_llm(prompt)
        try:
            return json.loads(response)
        except:
            continue
    return None


prompt = """
Extract name and age as JSON.

Text:
Bob is 40 years old
"""

print(ask_json(prompt))
```

## 🏗 Mini Project: Resume → JSON Parser

```python
# Input
resume_text = """
Rahul Sharma is a software engineer with 5 years of experience in Python and AI.
He lives in Bangalore.
"""
```

```python
# Prompt Design
prompt = f"""
Extract the following fields in JSON:
- name
- experience
- skills
- location

Rules:
- Output ONLY JSON
- No explanation

Text:
{resume_text}
"""
```

```python
# Output
result = ask_json(prompt)
print(result)
```

## 🧠 Key Learnings
- LLMs are probabilistic, not deterministic
- Prompt clarity = output quality
- JSON enforcement is fragile without retries
- You are “programming” using prompts
